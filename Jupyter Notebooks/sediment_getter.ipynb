```python
def get_sed_type(file):
    xl = pd.ExcelFile(file)
    xl = xl.parse()
    if 'LITHOLOGY' in xl:
        return 'good'        
    else:
        return 'bad'
```

```python
def get_sed():
    loc_or = 'C:\\Users\\mdulieu\\Documents\\hackforthesea\\Cod-Squad\\data\\sediment\\raw\\'

    files = [f for f in listdir(loc_or) if isfile(join(loc_or, f))]
    
    file_status = pd.DataFrame([])
       
    for xl_file in files:
        sed = get_sed_type(loc_or + xl_file)
        file_status = file_status.append(pd.DataFrame({'file_name':xl_file,'status':sed}, index=[0]), ignore_index=True)
        
    return file_status
```

```python
def fix_bad_file(master_file, bad_file):
    bad_class_cols = ['SHEPARD_CLASS',
        'CLASS',
        'SED_CLASS',
        'DOMINANT LITHOLOGY',
        'CLASSIFICATION',
        'DESCRIPTION',
        'TEXTURE',
        'SHEPARD_CL',
        'SHEPARD CLASSIFICATION',
        'Classification',
        'SEDCLASS',
        'SEDENV']
    
    for index, bad_file in bad_file.iterrows():
        file_path = loc_or + bad_file.file_name
        xl = pd.ExcelFile(file_path)
        xl = xl.parse()
        for col in bad_class_cols:
            if col in xl:
                xl['LITHOLOGY'] = xl[col]
                try:
                    master_file = master_file.append(xl[['LATITUDE', 'LONGITUDE', 'LITHOLOGY']])
                except: 
                    print(bad_file.file_name)
                    #exception for usgsecstdb
                    xl['LATITUDE'] = xl['Latitude']
                    xl['LONGITUDE'] = xl['Longitude']
                    master_file = master_file.append(xl[['LATITUDE', 'LONGITUDE', 'LITHOLOGY']])
                    pass
      
    return master_file
```

```python
def get_master_file():
    loc_or = 'C:\\Users\\mdulieu\\Documents\\hackforthesea\\Cod-Squad\\data\\sediment\\raw\\'
    
    master_file = pd.DataFrame()
    
    
#     good files processing
    good_files_mask = file_status['status'] == 'good'

    good_files = file_status[good_files_mask]

    for index, good_file in good_files.iterrows():
        file_path = loc_or + good_file.file_name
        xl = pd.ExcelFile(file_path)
        xl = xl.parse()
        master_file = master_file.append(xl[['LATITUDE', 'LONGITUDE', 'LITHOLOGY']])
        
#     bad files processing
    bad_files_mask = file_status['status'] == 'bad'

    bad_files = file_status[bad_files_mask]
    
    master_file = fix_bad_file(master_file, bad_files)
    
#     resetting the indexes    
    master_file = master_file.reset_index(drop=True)
        
    return master_file    
```

```python
def denom_litho():
    sediment_classes = [
        'GRAVEL',
        'SAND',
        'CLAY',
        'MUD',
        'SILT',
        'CLAY',
        'ROCK',
        'SEDIMENT',
        'BEDROCK'
    ] 

    sediment_classes_dict = {}

    for sediment_class in sediment_classes:
        sediment_classes_dict[sediment_class] = []
        #looking at sand denominations
        sediment_denom_full = pd.DataFrame()
        for index, row in master_file.iterrows():
            try:
                if row.LITHOLOGY.endswith(sediment_class):
                    sediment_denom_full = sediment_denom_full.append(row)
            except:
                pass
            
        sediment_classes_dict[sediment_class] = sediment_denom_full.LITHOLOGY.unique()

    return sediment_classes_dict
```

```python
def get_loc_denom(master_file):
    master_file['LITHO_GROUP'] =pd.Series()

    for index, row in master_file.iterrows():
        for sediment_class in sediment_classes_dict:
            try:
                #cleaning the litho string
                corrected_litho = row.LITHOLOGY.upper().replace(".", "")
                if corrected_litho.endswith(" "):
                    corrected_litho = corrected_litho[:-1]
            except:
                print('could not clean the following string, sorry')
                print(row.LITHOLOGY)
            #bucketing into a LITHO_GROUP category
            if corrected_litho in sediment_classes_dict[sediment_class]:
                master_file.loc[index,'LITHO_GROUP'] = sediment_class 
            else:
                pass

    return master_file
```
