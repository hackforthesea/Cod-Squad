```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
plt.rcParams['figure.figsize'] = (16,8)
```

```python
data = np.loadtxt('../data/bathymetric_data/GEBCO2014_-71.3956_41.3653_-65.8981_45.5522_30Sec_ESRIASCII.asc', skiprows=7)
```

```python
x = -71.395600000000 + np.arange(660) * 0.008340438247
y = 41.365300000000 + np.arange(502) * 0.008340438247
```

```python
cmap = mpl.colors.ListedColormap(['blue', 'cyan', 'black'])
bounds = [-600, -200, 0, 200]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)


fig, ax = plt.subplots()
ax.pcolormesh(y,x,data.T, cmap=cmap, norm=norm)
```
